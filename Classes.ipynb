Challenge 1 : Gestion d’un compte bancaire en POO

```python
class CompteBancaire :
    nom_proprietaire : str
    solde : float = 0.0

    def __init__(self , nom_proprietaire , solde = 0.0):
        self.nom_proprietaire = nom_proprietaire
        self.solde = solde 

    def deposer(self , montant) :
        self.solde += montant
        print(f"the amount {montant} added succesfully !")

    def retirer(self ,montant) : 
        
        if self.solde < montant :
            print(f"the solde in your account is not enough to retirer {montant}")
        else :
             self.solde -= montant
             print("the opiration is succusfulley completed !")
    
    def afficher_solde(self):
        return (f"your solde is {self.solde}")
```

Challenge 2 : Système de gestion d’école

```python
from abc import ABC , abstractmethod

class Person(ABC):
   
    def __init__(self , nom : str , prenom : str , age : int):
         self.nom = nom
         self.prenom = prenom
         self.age = age

    @abstractmethod
    def afficher_infos(self):
         pass

        
class Etudiant(Person) :
     
     def __init__(self , nom : str , prenom : str , age : int , matricule : str , notes : list[float] = None):
          super().__init__(nom, prenom, age)
          self.matricule = matricule
          self.notes = notes or []
     

     def ajouter_note(self , note : list[float] ):
          self.notes += note

     def moyenne(self) :
           if self.notes :
                return sum(self.notes) / len(self.notes)
           else : 
                return 0 
          
     
     def afficher_infos(self ):
          print(f" nom : {self.nom} \n prenom : {self.prenom} \n age : {self.age} \n  matricule : {self.matricule} \n notes : {self.notes}")

class Enseignant(Person) :
     
      def __init__(self , nom : str , prenom : str , age : int , specialite : str ,salaire : float):
           super().__init__(nom, prenom, age)
           self._specialite = specialite
           self._salaire = salaire

      def afficher_infos(self ):
         print(f" nom : {self.nom} \n prenom : {self.prenom} \n age : {self.age} \n specialite : {self._specialite} \n salaire : {self._salaire}")   

      @property 
      def specialite(self) :
           return self._specialite
      
      @specialite.setter
      def specialite(self, value) :
           self._specialite = value
    
      @property
      def salaire(self):
           return self._salaire
      
      @salaire.setter
      def salaire(self , value):
           self._salaire = value
           
           
class Ecole :

     def __init__(self , nom :str , liste_etudiants : list[Etudiant]  = None , liste_enseignants : list[Enseignant]  = None):
          self.nom = nom
          self.liste_etudiants = liste_etudiants or []
          self.liste_enseignants = liste_enseignants or []

     def ajouter_etudiant(self ,etudiant : Etudiant) :
          self.liste_etudiants.append(etudiant)
          
     def ajouter_enseignant(self ,enseignant : Enseignant) :
         self.liste_enseignants.append(enseignant)
          
     def afficher_tous_les_membres(self) :
          
          print("list of students : \n")
          for student in self.liste_etudiants : 
               student.afficher_infos()

          print("\nlist of enseignants : \n")
          for enseignant in self.liste_enseignants : 
               enseignant.afficher_infos()
```
